```python
import csv
dataset2_fixed = open("dataset2.csv","r")
#Memisahkan file menjadi beberapa baris
raw_lines = dataset2_fixed.readlines()
dataset2_fixed.close()
lines = [raw_line.replace("\n", "") for raw_line in raw_lines]

def convert_array_data_to_string(array_data):
    arr_cpy = array_data[:]
    for i in range(7):
        arr_cpy[i] = str(arr_cpy[i])
    return arr_cpy

def convert_line_to_data(line):
    raw_array_of_data = line.split(";")
    array_of_data = [data.strip() for data in raw_array_of_data]
    return array_of_data
```

```python
def tukar_nim(real_values):
    arr_tukar = real_values[:]
    for i in range(1,7):
        if len(arr_tukar[i])>3:
            arr_tukar[i], arr_tukar[0] = arr_tukar[0], arr_tukar[i]
    return arr_tukar

def nim_huruf(hasil_tukar_nim):
    arr_angka=hasil_tukar_nim[:]
    for i in range(7):
        arr_angka[i] = arr_angka[i].replace('B', '2')
        arr_angka[i] = arr_angka[i].replace('T', '7')
        arr_angka[i] = arr_angka[i].replace('S', '5')
        arr_angka[i] = arr_angka[i].replace('E', '3')
    return arr_angka

def angka_terisi(hasil_nim_huruf):
    arr_default = hasil_nim_huruf[:]
    for i in range(1,7):
        if (arr_default[i]== ''):
            arr_default[i] = arr_default[i].replace('', '50')
    return arr_default
```

```python
def nilai_default(hasil_nim_huruf):
    arr_default = hasil_nim_huruf[:]
    for i in range(1,7):
        arr_default[i]=int(arr_default[i])
        if(arr_default[i] < 0):
            arr_default[i] = 20
        elif(arr_default[i] > 100):
            arr_default[i] = 80
    return arr_default


def convert_nim(nilai_overlay):
    arr_nim = nilai_overlay[:]
    for i in range(7):
        if(i == 0):
            arr_nim[i] = str(arr_nim[i])
            if len(arr_nim[i])<10:
                arr_nim[i] = arr_nim[i]+'0'
                arr_nim[i] = int(arr_nim[i])
            elif len(arr_nim[i])>10:
                arr_nim[i]=str(arr_nim[i])
                arr_nim[i] = arr_nim[i][0:10]
                arr_nim[i] = int(arr_nim[i])
            else:
                arr_nim[i] = int(arr_nim[i])
    return arr_nim
```

```python
raw_header = lines.pop(0)
header = convert_line_to_data(raw_header)

datas =[]
for line in lines:
    array_of_data = convert_line_to_data(line)
    real_values = convert_array_data_to_string(array_of_data) 
    hasil_tukar_nim = tukar_nim(real_values)
    hasil_nim_huruf = nim_huruf(hasil_tukar_nim)
    hasil_angka_terisi = angka_terisi(hasil_nim_huruf)
    nilai_overlay = nilai_default(hasil_angka_terisi)
    hasil_nim = convert_nim(nilai_overlay)
    datas.append(hasil_nim)
print(hasil_nim)    
def convert_datas_to_string():
  string_data = ";".join(header) + "\n"
  for arr_data in datas:
    arr_data_all_string = [str(var) for var in arr_data]
    string_data += ";".join(arr_data_all_string)
    string_data += "\n"
  return string_data
datas_as_string = convert_datas_to_string()
# print(datas_as_string)
f = open("dataset2_fixed.csv", "x")
f.write(datas_as_string)
f.close()
```

```python
def nilai_max(value):
    return max(value)

def nilai_min(value):
    return min(value)
    
def Modus(value):
    return(max(value, key = value.count))
    
def Median(value):
    value = sorted(value)
    n = len(value)

    if n % 2 == 1:
        return value[n//2]
    else:
        x = n//2
        return (value[x]+value[x+1])/2

def Mean(data):
    return round(sum(data)/len(data),3)
```

```python
def variance(value):
    data = []

    for i in range(len(value)):
        data.append((value[i] - Mean(value)) ** 2)
    return sum(data)/(len(data)-1)

def standar_derivative(value):
    std_dev = variance(value)**0.5
    return std_dev
```

```python
def quartile1(value):
    m = 3*(len(value)+1)/4
    return m
    
def quartile3(value):
    n = 3*(len(value)+1)/4
    return n

def IQR(value):
    return quartile3(value)-quartile1(value)

def SIQR(value):
    return IQR(value)/2


def one_Step(value):
    return IQR(value)*1.5
```
